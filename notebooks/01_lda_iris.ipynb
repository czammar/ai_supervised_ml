```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

# 1. Cargar datos
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
colors = ['navy', 'turquoise', 'darkorange']

# 2. Escalar datos (LDA tambin se beneficia de la estandarizacin)
X_scaled = StandardScaler().fit_transform(X)

# 3. Aplicar LDA (mximo n_components = n_classes - 1)
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_scaled, y) # LDA es supervisado, requiere 'y'

# 4. Visualizacin en 2D
plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1],
                alpha=.8, color=color, label=target_name, edgecolors='k')

plt.title('LDA de Iris: Proyeccin en 2D (Mximo de componentes)')
plt.xlabel(f'LD1 ({lda.explained_variance_ratio_[0]:.2%} varianza)')
plt.ylabel(f'LD2 ({lda.explained_variance_ratio_[1]:.2%} varianza)')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.show()
```
