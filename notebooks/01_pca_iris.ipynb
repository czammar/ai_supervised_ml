```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# 1. Cargar los datos de Iris

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

colors = ['navy', 'turquoise', 'darkorange']
```

```python
pd.DataFrame(X, columns=iris.feature_names)
```

```python
y
```

```python
# Escalamos los datos antes de aplicar PCA
X_scaled = StandardScaler().fit_transform(X)
```

```python
# --- VISUALIZACIÓN 2D ---
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_pca_2d[y == i, 0],
        X_pca_2d[y == i, 1],
        color=color,
        alpha=.8,
        lw=2,
        label=target_name
        )

plt.title('PCA de Iris: Proyección en 2D')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.2%} var.)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.2%} var.)')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.grid(alpha=0.3)
plt.show()
```

```python
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection='3d')

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax.scatter(X_pca_3d[y == i, 0], X_pca_3d[y == i, 1], X_pca_3d[y == i, 2],
               color=color, s=40, label=target_name, edgecolors='k')

ax.set_title('PCA de Iris: Proyección en 3D')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.legend()
plt.show()
```
