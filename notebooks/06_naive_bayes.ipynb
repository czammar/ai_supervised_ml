## Naive Bayes

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
wine = load_wine()
X, y = wine.data, wine.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 3. Entrenamiento del modelo
model = GaussianNB()
model.fit(X_train, y_train)

# 4. Predicción
y_pred = model.predict(X_test)
```

```python
print("ESTADÍSTICAS DEL MODELO NAIVE BAYES")
print("-" * 40)
# Imprime Precisión, Recall y F1-Score por cada tipo de vino
print(classification_report(y_test, y_pred, target_names=wine.target_names))
```

```python
# Creación del Plot de la Matriz de Confusión
fig, ax = plt.subplots(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=wine.target_names)

# Personalización visual
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.title("Matriz de Confusión: Clasificación de Vinos (Naive Bayes)")
plt.show()
```
